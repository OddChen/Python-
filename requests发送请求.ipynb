```python
import requests

url = "http://dianping.com"
header = {
    "Host": "www.dianping.com",
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.138 Safari/537.36"
}
res = requests.get(url, headers=header)#发起请求
print(res.encoding)#查看编码
print(res.headers)#里面 如果没有Content-Type encoding=utf-8 如果charset，就以设置的为准，否则就是iso8859-1
print(res.status_code)#获取状态码
#html = res.text
#print(html)
res.encoding = "UTF-8"
html = res.text
print(html)
```
